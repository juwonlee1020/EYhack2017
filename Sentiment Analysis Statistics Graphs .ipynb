```python
#import statements
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sentiment_ksdf = pd.read_csv("kickstarter_data.csv")
```

```python
sentiment_ksdf.head()
```

```python
type(sentiment_ksdf["sentiment_score"])
positive_sentiment_list = []
for sentiment in sentiment_ksdf["sentiment_score"]:
    if sentiment > 0.2:
        positive_sentiment_list.append(True)
    else:
        positive_sentiment_list.append(False)
positive_sentiment_list
```

```python
sentiment_ksdf["positive_sentiment"] = positive_sentiment_list
```

```python
sentiment_ksdf.head()
```

```python
#Not a very clearn pattern
#The actual sen
sns.countplot(x="success", hue="positive_sentiment", data=sentiment_ksdf)
```

```python
sns.boxplot(x="success", y="sentiment_score", data=sentiment_ksdf, palette="rainbow"
```

```python
plt.ylim(0,80)
sns.boxplot(x="success", y="sentiment_magnitude", data=sentiment_ksdf, palette="rainbow")
```
